```python
import spacy

# Load English NLP model
nlp = spacy.load("en_core_web_sm")

# Load your text
with open("/content/How to set up an educational YouTube channel #youtubechannel #youtubetips.txt", "r") as file:
    text = file.read()

# Segment text into sentences
doc = nlp(text)
action_steps = []

# Identify sentences with action verbs
for sentence in doc.sents:
    if any(token.pos_ == "VERB" for token in sentence):
        action_steps.append(sentence.text.strip())

# Display the action steps
print("PlAN OF ACTION:")
for i, action in enumerate(action_steps, 1):
    print(f"Step {i}: {action}")
```
